**Explore Data**

---



```python
from matplotlib.font_manager import dataclasses
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Age distribution
plt.subplot(2, 3, 1)
sns.histplot(data['age'], kde=True, bins=30)
plt.title('Age Distribution')

# Distribution of Resting Blood Pressure
plt.subplot(2, 3, 2)
sns.histplot(data['trtbps'], kde=True, bins=30, color='skyblue')
plt.title('Resting Blood Pressure Distribution')

# Distribution of Cholesterol Level
plt.subplot(2, 3, 3)
sns.histplot(data['chol'], kde=True, bins=30, color='salmon')
plt.title('Cholesterol Level Distribution')

# Distribution of Maximum Heart Rate Achieved
plt.subplot(2, 3, 4)
sns.histplot(data['thalachh'], kde=True, bins=30, color='green')
plt.title('Max Heart Rate Distribution')

# Distribution of ST Depression Induced by Exercise Relative to Rest
plt.subplot(2, 3, 5)
sns.histplot(data['oldpeak'], kde=True, bins=30, color='purple')
plt.title('ST Depression Distribution')

# Distribution of Target Variable (Heart Attack)
plt.subplot(2, 3, 6)
sns.countplot(x='output', data=data, palette="Set2")
plt.title('Heart Attack Distribution (0 = No, 1 = Yes)')

plt.tight_layout()
plt.show()
```

```python
# Set up the figure and axis
plt.figure(figsize=(10, 8))

# Relationship between Age and Heart Attack
plt.subplot(2, 3, 1)
sns.boxplot(x='output', y='age', data=data, palette="coolwarm")
plt.title('Age vs Heart Attack')

# Relationship between Resting Blood Pressure and Heart Attack
plt.subplot(2, 3, 2)
sns.boxplot(x='output', y='trtbps', data=data, palette="coolwarm")
plt.title('Resting Blood Pressure vs Heart Attack')

# Relationship between Cholesterol Level and Heart Attack
plt.subplot(2, 3, 3)
sns.boxplot(x='output', y='chol', data=data, palette="coolwarm")
plt.title('Cholesterol Level vs Heart Attack')

# Relationship between Max Heart Rate and Heart Attack
plt.subplot(2, 3, 4)
sns.boxplot(x='output', y='thalachh', data=data, palette="coolwarm")
plt.title('Max Heart Rate vs Heart Attack')

# Relationship between ST Depression and Heart Attack
plt.subplot(2, 3, 5)
sns.boxplot(x='output', y='oldpeak', data=data, palette="coolwarm")
plt.title('ST Depression vs Heart Attack')

# Relationship between Sex and Heart Attack
plt.subplot(2, 3, 6)
sns.countplot(x='sex', hue='output', data=data, palette="Set2")
plt.title('Sex vs Heart Attack (0 = Female, 1 = Male)')

plt.tight_layout()
plt.show()
```

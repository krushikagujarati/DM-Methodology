In this step, we'll explore the dataset using visualizations and statistics to understand the characteristics and distribution of the data. We'll start by looking at the distribution of scores and then explore some relationships between the variables.

```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
from google.colab import files

uploaded = files.upload()
import pandas as pd
data = pd.read_csv('exams.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set_style("whitegrid")

# Create a figure with subplots
fig, ax = plt.subplots(3, 1, figsize=(5, 7))

# Plot distribution of math, reading, and writing scores
sns.histplot(data['math score'], kde=True, ax=ax[0], color="skyblue", bins=30)
ax[0].set_title('Distribution of Math Scores')
ax[0].set_xlabel('Math Score')
ax[0].set_ylabel('Count')

sns.histplot(data['reading score'], kde=True, ax=ax[1], color="salmon", bins=30)
ax[1].set_title('Distribution of Reading Scores')
ax[1].set_xlabel('Reading Score')
ax[1].set_ylabel('Count')

sns.histplot(data['writing score'], kde=True, ax=ax[2], color="green", bins=30)
ax[2].set_title('Distribution of Writing Scores')
ax[2].set_xlabel('Writing Score')
ax[2].set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

The histograms show the distributions of scores in math, reading, and writing exams. Here are some observations:

1. Math Scores: The distribution is slightly left-skewed with a majority of students scoring between 60 and 80. There's also a small peak around the 90-100 range, suggesting a group of high achievers in math.
2. Reading Scores: This distribution is approximately normal, with the majority of students scoring between 60 and 80.
3. Writing Scores: Similar to the reading scores, the distribution is approximately normal with a central peak around 70.

```python
# Create a figure with subplots
fig, ax = plt.subplots(3, 1, figsize=(7, 10))

# Ordering for the parental level of education for better visualization
order_education = [
    'some high school',
    'high school',
    'some college',
    'associate\'s degree',
    'bachelor\'s degree',
    'master\'s degree'
]

# Plot average scores by parental level of education
sns.barplot(x='parental level of education', y='math score', data=data, ax=ax[0], palette="pastel", order=order_education)
ax[0].set_title('Average Math Scores by Parental Level of Education')
ax[0].set_ylabel('Average Math Score')
ax[0].tick_params(axis='x', rotation=30)

sns.barplot(x='parental level of education', y='reading score', data=data, ax=ax[1], palette="pastel", order=order_education)
ax[1].set_title('Average Reading Scores by Parental Level of Education')
ax[1].set_ylabel('Average Reading Score')
ax[1].tick_params(axis='x', rotation=30)

sns.barplot(x='parental level of education', y='writing score', data=data, ax=ax[2], palette="pastel", order=order_education)
ax[2].set_title('Average Writing Scores by Parental Level of Education')
ax[2].set_ylabel('Average Writing Score')
ax[2].tick_params(axis='x', rotation=30)

# Adjust layout
plt.tight_layout()
plt.show()

```

From the bar plots comparing exam scores by the parental level of education, we observe:

1. Math Scores: Students whose parents have a master's degree tend to have the highest average math scores. Those with parents who attended "some high school" or "high school" have the lowest average scores.
2. Reading Scores: A similar trend is observed in reading scores. Students with parents holding master's degrees outperform others. The scores generally increase with the increasing level of parental education.
3. Writing Scores: This trend persists in writing scores as well. Students with parents having higher educational qualifications tend to score better on average.

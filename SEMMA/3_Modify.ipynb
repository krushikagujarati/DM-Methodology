In this step, we'll pre-process and clean the dataset to make it suitable for modeling. This includes handling missing values (if any), encoding categorical variables, and normalizing numerical variables. Let's start by checking for any missing values.

```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
from google.colab import files

uploaded = files.upload()
import pandas as pd
data = pd.read_csv('exams.csv')
```

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()

missing_values
```

```python
# One-hot encode the categorical variables
data_encoded = pd.get_dummies(data, columns=[
    'gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'
], drop_first=True)

data_encoded.head()
```

The categorical variables have been successfully one-hot encoded, resulting in binary columns for each category. For instance:

* The "gender" column is encoded as gender_male where 1 represents male and 0 represents female.
* The "race/ethnicity" column is split into multiple columns such as race/ethnicity_group B, race/ethnicity_group C, etc., where 1 indicates the presence of the category and 0 indicates its absence.

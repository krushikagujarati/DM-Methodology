```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
from google.colab import files

uploaded = files.upload()


import pandas as pd
data = pd.read_csv('House_Rent_Dataset.csv')
```

```python
def extract_total_floors(x):
    if ' out of ' in x:
        return int(x.split(' out of ')[1])
    return None

data['Total Floors'] = data['Floor'].apply(extract_total_floors)

# Display the first few rows to verify the changes
data.head()
```

```python
import matplotlib.pyplot as plt

# List of columns to explore
columns_to_explore = ["BHK", "Rent", "Size", "Total Floors", "Bathroom"]

# Plot histograms for each column
plt.figure(figsize=(10, 8))

for idx, col in enumerate(columns_to_explore, 1):
    plt.subplot(2, 3, idx)
    data[col].hist(bins=30, edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
